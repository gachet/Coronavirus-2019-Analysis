```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data = pd.read_csv('dataset/covid_19_data.csv')
line_list = pd.read_csv('dataset/COVID19_line_list_data.csv')
open_line_list = pd.read_csv('dataset/COVID19_open_line_list.csv')
confirmed = pd.read_csv('dataset/time_series_covid_19_confirmed.csv')
deaths = pd.read_csv('dataset/time_series_covid_19_deaths.csv')
recovered = pd.read_csv('dataset/time_series_covid_19_recovered.csv')
```

```python
datetime.datetime.strptime('1/22/2020 17:00', '%m/%d/%Y %H:%M')
```

```python
data.tail()
```

```python
data2 = data[['ObservationDate', 'Confirmed', 'Deaths', 'Recovered']]
```

```python
data3 = data2.groupby('ObservationDate').sum()
```

```python
data3['ObservationDateCol'] = data3.index
```

```python
data3['ObservationDateCol'] = data3['ObservationDateCol'].apply(lambda d: datetime.datetime.strptime(d, '%m/%d/%Y'))
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data3['ObservationDateCol'], data3['Confirmed'])
sns.lineplot(data3['ObservationDateCol'], data3['Deaths'])
sns.lineplot(data3['ObservationDateCol'], data3['Recovered'])
plt.legend(['Confirmed', 'Deaths', 'Recovered'])
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data3['ObservationDateCol'], data3['Deaths']*100/data3['Confirmed'])
plt.title('Mortality rate')
```
